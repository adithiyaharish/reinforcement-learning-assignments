## Exercise 1.1: Array Creation and Basic Operations

```python
array_1D = np.arange(10, 51, 5) # Creates a 1D array starting from 0 ending at 50 with a step size of 5

print("Shape:", array_1D.shape)
print("Size:", array_1D.size)
print("Data type:", array_1D.dtype)

sum = array_1D.sum()
max = array_1D.max()
max_index = np.where(array_1D == max)[0][0] # Returns the index position of the first encountered max_value

print("Sum of elements:", sum)
print("Maximum value:", max)
print("Index of maximum value:", max_index)
```

```python
array1 = np.array([[10],
                   [20],
                   [30]])
array2 = np.array([1, 2, 3, 4])

print("Broadcasted Array:")
print(array1 + array2)
```

Shape of `array1` : `3x1`

Shape of `array2` : `1x4`

Broadcasting `array1` and `array2` by adding them up creates a new array of size `3x4` similar to matrix multiplication.

This is done by stretching `array1` to create 4 columns and stretching `array2` to create 3 rows of the same values and then the addition takes place.

## Exercise 2.2: Conditional Operations and Boolean Indexing

```python
np.random.seed(0)
array_5x5 = np.random.randint(1, 101, size = (5, 5)) # Creates a 5x5 matrix with random integer elements
```



---



```python
print("Elements greater than 50:")
print(array_5x5 > 50) # Boolean indexing for elements greater than 50
print("Elements:", array_5x5[array_5x5 > 50])
```

```python
even_numbers = (array_5x5 % 2 == 0) # Creates a 1D array of all even numbers

print("Array:")
print(array_5x5)

array_5x5[even_numbers] = -array_5x5[even_numbers] # Reassigns to a negative value

print("Array with even numbers replaced by their negative values:")
print(array_5x5)

# Changing it back to original for the next few questions
array_5x5[even_numbers] = -array_5x5[even_numbers]
```

```python
print("Number of elements divisible by 3:", np.sum(array_5x5 % 3 == 0))
```

```python
new_array = np.where(array_5x5 > 75, 1,
                     np.where(array_5x5 < 25, -1, 0))

print("Original Array:\n", array_5x5)
print("\nNew Array (values > 75 are 1, < 25 are -1, others are 0):\n", new_array)
```

## Exercise 3.2: Statistical Visualizations

```python
data = np.random.normal(loc = 0, scale = 1, size = 1000) # Random data for normal with mean = 0 and variance = 1

count, bins, _ = plt.hist(data, bins = 30, density = True, alpha = 0.6, color = 'pink', edgecolor = 'black', label = "Histogram")

bin_centers = 0.5 * (bins[1:] + bins[:-1])
hist_density = count

plt.plot(bin_centers, hist_density, color = 'red', linewidth = 2, label = 'Density Curve')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram')
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(0)

group1 = np.random.normal(loc = 50, scale = 10, size = 100) # Data for mean = 50
group2 = np.random.normal(loc = 60, scale = 10, size = 100) # Data for mean = 60
group3 = np.random.normal(loc = 70, scale = 10, size = 100) # Data for mean = 70
group4 = np.random.normal(loc = 80, scale = 10, size = 100) # Data for mean = 80

scatter = [group1, group2, group3, group4]


boxprops = dict(facecolor = 'skyblue', color = 'blue') # Properties of the boxes
medianprops = dict(color = 'red', linewidth = 2) # Properties of the median line
flierprops = dict(marker = 'o', color = 'gray', alpha = 0.5) # Properties of the fliers

plt.boxplot(scatter, patch_artist = True, boxprops = boxprops, medianprops = medianprops, flierprops = flierprops)

plt.xlabel('Groups')
plt.ylabel('Values')
plt.xticks([1, 2, 3, 4], ['Group A', 'Group B', 'Group C', 'Group D'])
plt.title("Box Plot of 4 Groups with Different Means")
plt.ylabel("Values")
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.show()
```

## Exercise 3.3: Advanced Plotting

```python
from mpl_toolkits.mplot3d import Axes3D
import random

def generate_random_list(start, stop, step, num_elements):
    '''
    Generates a list of random numbers within a specified range with step sizes.
    '''
    random_list = []
    for _ in range(num_elements):
        random_num = random.randrange(start, stop, step)
        random_list.append(random_num)
    return random_list

fig = plt.figure(figsize=(12, 10))

# Figure 1
ax1 = fig.add_subplot(2, 2, 1)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = generate_random_list(50, 250, 10, len(months))
ax1.bar(months, sales, color='skyblue')
ax1.set_title("Monthly Sales Data")
ax1.set_xlabel("Month")
ax1.set_ylabel("Sales")

# Figure 2
ax2 = fig.add_subplot(2, 2, 2)
labels = ['A', 'B', 'C', 'D', 'E']
sizes = generate_random_list(10, 40, 5, len(labels))
colors = plt.cm.Set3(np.linspace(0, 1, 5)) # Sets color for each label
ax2.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors) # Plots a pie chart with the data
ax2.set_title("Market Share")

# Figure 3
ax3 = fig.add_subplot(2, 2, 3)
data_heatmap = np.random.rand(100, 5)
corr = np.corrcoef(data_heatmap.T) # Pearson correlation coefficient for the correlation matrix
heatmap = ax3.imshow(corr, cmap='coolwarm', interpolation='nearest')
ax3.set_title("Correlation Matrix", fontsize=12)
ax3.set_xticks(range(5))
ax3.set_yticks(range(5))
ax3.set_xticklabels([f'{i+1}' for i in range(5)])
ax3.set_yticklabels([f'{i+1}' for i in range(5)])
plt.colorbar(heatmap, ax=ax3)

# Figure 4
ax4 = fig.add_subplot(2, 2, 4, projection='3d') # 3d projection
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y) # Z = f(x, y) = sin(x) * cos(y)
surface = ax4.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax4.set_title("3D Surface: sin(x) * cos(y)", fontsize=12)
ax4.set_xlabel("X-axis")
ax4.set_ylabel("Y-axis")
ax4.set_zlabel("Z-axis")

plt.tight_layout()
plt.subplots_adjust(wspace=0.3, hspace=0.4)
plt.show()
```

# **Bonus Section: Real-World Application**

```python
days = np.arange(365) # List of all days in the year

base_temp = 25 + 10 * np.sin(2 * np.pi * days / 365) # Base temperature around 25°C

noise = np.random.normal(0, 2, 365) # Random noise ranging from 0-2

temps = base_temp + noise

extreme_days = np.random.choice(days, size = 5, replace = False) # Random outliers
temps[extreme_days] += np.random.choice([10, -10], size = 5)

months = np.array([31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31])
month_indices = np.split(days, np.cumsum(months)[:-1])
monthly_averages = np.array([temps[i].mean() for i in month_indices])

hottest_day = np.argmax(temps) # Calculates the hottest day (maximum temperature)
coldest_day = np.argmin(temps) # Calculates the coldest day (minimum temperature)

moving_avg = np.convolve(temps, np.ones(7)/7, mode='valid')

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(days, temps, label = "Daily Temp", alpha = 0.6)
plt.plot(days[3:-3], moving_avg, color = 'red', label = "7-Day Moving Avg", linewidth = 2)
plt.axvline(hottest_day, color = 'orange', linestyle = '--', label = "Hottest")
plt.axvline(coldest_day, color = 'blue', linestyle = '--', label = "Coldest")
plt.title("Daily Temperatures Over a Year")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
data_monthly = [temps[i] for i in month_indices]
plt.boxplot(data_monthly, tick_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title("Monthly Temperature Distribution")
plt.ylabel("Temperature (°C)")

plt.subplot(2, 2, 3)
plt.hist(temps, bins = 30, color = 'skyblue', edgecolor = 'black')
plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(np.arange(1, 13), monthly_averages, marker = 'o', color = 'green')
plt.title("Monthly Average Temperatures")
plt.xlabel("Month")
plt.ylabel("Average Temp (°C)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
